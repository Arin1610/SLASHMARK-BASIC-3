```python
import random
import string

def generate_password(length=12):
    # Define the character set for generating the password
    characters = string.ascii_letters + string.digits + string.punctuation

    # Generate password using random.choices() method
    password = ''.join(random.choices(characters, k=length))
    
    return password

def main():
    print("Welcome to the Password Generator!")
    length = int(input("Enter the length of the password you want to generate: "))
    
    password = generate_password(length)
    print("Generated Password:", password)

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import string
import random

class PasswordGeneratorGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("Password Generator")

        self.label_length = tk.Label(master, text="Password Length:")
        self.label_length.grid(row=0, column=0, padx=10, pady=5, sticky=tk.W)

        self.entry_length = tk.Entry(master)
        self.entry_length.grid(row=0, column=1, padx=10, pady=5, sticky=tk.W)

        self.label_options = tk.Label(master, text="Options:")
        self.label_options.grid(row=1, column=0, padx=10, pady=5, sticky=tk.W)

        self.include_special = tk.BooleanVar()
        self.check_special = tk.Checkbutton(master, text="Special Characters", variable=self.include_special)
        self.check_special.grid(row=1, column=1, padx=10, pady=5, sticky=tk.W)

        self.include_numbers = tk.BooleanVar()
        self.check_numbers = tk.Checkbutton(master, text="Numbers", variable=self.include_numbers)
        self.check_numbers.grid(row=2, column=1, padx=10, pady=5, sticky=tk.W)

        self.include_uppercase = tk.BooleanVar()
        self.check_uppercase = tk.Checkbutton(master, text="Uppercase Letters", variable=self.include_uppercase)
        self.check_uppercase.grid(row=3, column=1, padx=10, pady=5, sticky=tk.W)

        self.button_generate = tk.Button(master, text="Generate Password", command=self.generate_password)
        self.button_generate.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

        self.label_generated_password = tk.Label(master, text="")
        self.label_generated_password.grid(row=5, column=0, columnspan=2, padx=10, pady=5)

    def generate_password(self):
        length = self.entry_length.get()
        if not length.isdigit():
            messagebox.showerror("Error", "Please enter a valid number for password length.")
            return

        length = int(length)
        if length <= 0:
            messagebox.showerror("Error", "Password length should be greater than zero.")
            return

        characters = string.ascii_letters
        if self.include_special.get():
            characters += string.punctuation
        if self.include_numbers.get():
            characters += string.digits
        if self.include_uppercase.get():
            characters += string.ascii_uppercase

        if not characters:
            messagebox.showerror("Error", "Please select at least one option for password generation.")
            return

        generated_password = ''.join(random.choice(characters) for _ in range(length))
        self.label_generated_password.config(text=f"Generated Password: {generated_password}")

def main():
    root = tk.Tk()
    password_generator_gui = PasswordGeneratorGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
